```python
import pandas as pd
import MySQLdb as mysql

weather=pd.read_csv('weather.csv')
shark=pd.read_csv('attacks_new.csv')

pd.set_option('display.max_columns', 50)
```

```python
weather['country']=weather['NAME'].str.extract(r'^.+,\s(.+)')
weather['country']=weather['country'].str.replace('SF', 'South Africa')
weather['country']=weather['country'].str.replace('NC US','North Carolina')
weather['country']=weather['country'].str.replace('FL US','Florida')
weather['country']=weather['country'].str.replace('AS','Australia')
```

```python
weather['DATE']=pd.to_datetime(weather['DATE'])
```

```python
shark['Date']=pd.to_datetime(shark['Date'],errors = 'coerce')
```

```python
def query(s):
    try:
        conn = mysql.connect('shark','weather')
        data = pd.io.sql.read_sql(s,conn)
        data.index += 1
        return data
    except mysql.Error as e:
        print(e)
    finally:
        if conn:
            conn.close()
```

```python
q='''
SELECT *
FROM shark weather
WHERE country = "USA"
'''

query(q)
```
