```python
from bokeh.plotting import figure, output_notebook, show
from bokeh.models import ColumnDataSource
from bokeh.models.tools import HoverTool
from bokeh.palettes import Spectral5
from bokeh.transform import factor_cmap
from bokeh.palettes import Category20

output_notebook()
```

```python
group=shark.groupby('Activity_type').count()
group
#missing value doesn't have any data in the row so delete this row completely when showing the graph
```

```python
#getting rid of missing value activity
group = group[group.index != 'nan']
group
```

```python
group=shark.groupby('Activity_type').count().sort_values(by = ['Year'], ascending = False)
group.index = [x.title() for x in group.index]  #upper case index
group.index.rename('Activity_type', inplace=True)
group
```

```python
from bokeh.palettes import Category20 as palette
import itertools
from bokeh.palettes import all_palettes
from bokeh.palettes import Viridis256

color=Viridis256[:256:11]

source=ColumnDataSource(group)
Activities=source.data['Activity_type'].tolist()


s = figure(x_range=Activities)
#color_map = factor_cmap(field_name='COUNTRY_FLYING_MISSION', 
                  #  palette=Spectral5, factors=countries)
    
color_map = factor_cmap(field_name='Activity_type', 
                    palette=color, factors=Activities)

#s.vbar(x='COUNTRY_FLYING_MISSION', top='TOTAL_TONS', source=shark, width=0.70, color=color_map)

s.vbar(x='Activity_type', top='Year', source=source ,width=.7, color=color_map)

s.title.text ='Global Shark Attack by Activity'
s.xaxis.axis_label = 'Activity'
s.yaxis.axis_label = 'Attacks'
#s.xaxis.major_label_orientation = "vertical"

s.xaxis.major_label_orientation = np.pi/2

hover = HoverTool()
hover.tooltips = [
    ("Activity", "@Activity_type"),
     ("Attacks", "@Year")]

hover.mode = 'vline'

s.add_tools(hover)

show(s)
```

```python
group=shark.groupby(['Country','Activity_type'], as_index= False).count().sort_values(by = ['Year'], ascending = False).head(20)

from bokeh.models import ColumnDataSource, FactorRange

group=shark.groupby(['Country','Activity_type'], as_index= False).count().sort_values(by = ['Year'], ascending = False).head(20)

color=Viridis256[:256:10]

country = list(group.Country)
activity = list(group.Activity_type)

x = [(country[x],activity[x]) for x in range(len(country))]
count =  list(group.Year)

source=ColumnDataSource(data=dict(x=x, count=count, country=country, activity=activity))

s = figure(x_range=FactorRange(*x))
    
color_map = factor_cmap(field_name='x', 
                    palette=color, factors=x)

#s.vbar(x='COUNTRY_FLYING_MISSION', top='TOTAL_TONS', source=shark, width=0.70, color=color_map)

s.vbar(x='x', top='count', source=source ,width=.1, color=color_map)

s.title.text ='Global Shark Attack by Country & Activity'
s.xaxis.axis_label = 'Country and Activity'
s.yaxis.axis_label = 'Attacks'
#s.xaxis.major_label_orientation = "vertical"

s.xaxis.major_label_orientation = np.pi/2

hover = HoverTool()
hover.tooltips = [
    ("Country", "@country"),
    ("Activity", "@activity"),
     ("Attacks", "@count")]

hover.mode = 'vline'

s.add_tools(hover)

show(s)
```
